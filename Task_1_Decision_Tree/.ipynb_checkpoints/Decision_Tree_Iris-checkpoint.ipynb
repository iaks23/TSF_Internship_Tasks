### Building a Classification Model

```python
#Dropping the ID column since it is not of much use.
iris_data.drop(['Id'], axis=1, inplace=True)
```

```python
iris_data.info()
```

```python
#Encoding our data so that, prediction becomes easy. Column 'Species' is of type Object, 
#We apply Label Encoding on the target column

# Import label encoder 
from sklearn import preprocessing 

# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 

# Encode labels in column 'Species'. 
iris_data['Species']= label_encoder.fit_transform(iris_data['Species']) 
```

```python
#Defining feature & target variables
X = iris_data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = iris_data['Species']

#Splitting data into test & train set
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
iris_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)
iris_gini.fit(Xtrain, Ytrain)
print('Decision Tree Created & Trained')
```

### Predicting test data and comparing

```python
Ypred = iris_gini.predict(Xtest)
```

```python
compare_results = pd.DataFrame({'Actual Data': Ytest, 'Predicted Data': Ypred})
compare_results.head(10)
```

### Visualizing the Decision Tree 

```python
fn = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
cn = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(iris_gini,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('Iris_Dst.png')
```

### Checking the accuracy of our model

```python
from sklearn.metrics import accuracy_score
```

```python
print('The accuracy on unseen test data is: ', accuracy_score(Ytest, Ypred)*100,'%')
```
